```python
def organize_articles(data):
    articles = []
    current_article = None

    for entry in data:
        heading = entry.get("heading")
        body = entry.get("body")
        
        # If the heading is identified as a new article title
        if current_article is None or not current_article["documents"]:
            current_article = {
                "Title": heading,
                "documents": [{"body": body}]
            }
            articles.append(current_article)
        elif heading.startswith("Main") or heading.startswith("Another"):  # Detect new article title
            # Start a new article section
            current_article = {
                "Title": heading,
                "documents": [{"body": body}]
            }
            articles.append(current_article)
        else:
            # Add the section to the current article
            current_article["documents"].append({
                "heading": heading,
                "body": body
            })

    return articles
```

```python
def structure_article_content(url_list):
    
    try:
        for url in url_list:
            response = requests.get(url)
            response.raise_for_status()
            soup = BeautifulSoup(response.text, 'html.parser')
            
            articles = []  # List to store multiple articles if any
            
            # Find all article sections or main content areas
            main_sections = soup.find('div', class_='content-main mb-5')
            
            for section in main_sections:
                article_dict = {}
                documents = []
                current_document = {}
                
                # Find the main title (h1)
                title = section.find('h1')
                if title:
                    article_dict["Title"] = title
                
                # Find all headings and paragraphs
                elements = main_sections.find_all(['h1', 'h2', 'p'])
                
                for element in elements:
                    if element.name == 'h1':
                        # Get paragraphs after h1
                        next_element = element.find_next_sibling()
                        while next_element and next_element.name == 'p':
                            current_document = {
                                "body": next_element.text.strip(),
                            }
                            documents.append(current_document)
                            next_element = next_element.find_next_sibling()
                            
                    elif element.name == 'h2':
                        # Start new document with h2
                        current_document = {
                            "heading": element.text.strip(),
                        }
                        # Get paragraphs after h2
                        next_element = element.find_next_sibling()
                        while next_element and next_element.name == 'p':
                            current_document["body"] = next_element.text.strip()
                            next_element = next_element.find_next_sibling()
                        documents.append(current_document)
                
                article_dict["documents"] = documents
                articles.append(article_dict)
        
        return articles

    except requests.RequestException as e:
        print(f"Error fetching the webpage: {e}")
        return None
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_article(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    current_article = None
    documents = []

    # Step 1: Identify all <h1> and <h2> headings and their following paragraphs
    h1 = soup.find('h1')  # Assuming one <h1> per article
    if h1:
        article_title = h1.get_text(strip=True)
        current_article = {
            "Title": article_title,
            "documents": []
        }
        
        # Get paragraphs after the <h1>
        paragraphs_after_h1 = []
        for sibling in h1.find_next_siblings():
            if sibling.name == 'p':
                paragraphs_after_h1.append(sibling.get_text(strip=True))
            elif sibling.name == 'h2':
                # When a new <h2> is encountered, stop collecting paragraphs for <h1>
                break
        
        # Add the paragraphs after <h1> as a part of the document
        if paragraphs_after_h1:
            current_article["documents"].append({
                "body": " ".join(paragraphs_after_h1),
            })

        # Step 2: Find all <h2> and its subsequent paragraphs
        for h2 in soup.find_all('h2'):
            h2_text = h2.get_text(strip=True)
            paragraphs_after_h2 = []

            # Collect paragraphs after each <h2>
            for sibling in h2.find_next_siblings():
                if sibling.name == 'p':
                    paragraphs_after_h2.append(sibling.get_text(strip=True))
                elif sibling.name in ['h1', 'h2']:  # Stop at the next heading
                    break

            if paragraphs_after_h2:
                current_article["documents"].append({
                    "heading": h2_text,
                    "body": " ".join(paragraphs_after_h2)
                })

    return current_article

def scrape_multiple_articles(urls):
    articles = []
    
    for url in urls:
        article_data = scrape_article(url)
        if article_data:
            articles.append(article_data)
    
    return articles


# Scrape and organize multiple articles
articles = scrape_multiple_articles(url_list)

# Print the structured articles
import pprint
pprint.pprint(articles)
```
